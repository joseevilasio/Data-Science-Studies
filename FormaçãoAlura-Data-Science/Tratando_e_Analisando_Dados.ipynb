## Identificando e Removendo Outliers

```python
dados_aluguel_v3 = pd.read_csv('./assets/02-Curso/1-aula/aluguel_residencial_v3.csv', sep = ';')
#dados_aluguel_selecao.boxplot(['Valor'], by = ['Tipo'])
dados_aluguel_v3.head(10)
```

```python
dados_amostra.boxplot(['Valor'], by = ['Tipo'])
```

```python
dados = dados_amostra
grupo_tipo = dados.groupby("Tipo")["Valor"]
grupo_tipo.head(10)
```

```python
valor = grupo_tipo
Q1 = valor.quantile(.25)
Q3 = valor.quantile(.75)
IIQ = Q3 - Q1
limite_inferior = Q1 - 1.5 * IIQ
limite_superior = Q3 + 1.5 * IIQ
```

```python
for tipo in grupo_tipo.groups.keys():
    print(tipo)
```

```python
dados_aluguel_v4 = pd.DataFrame()
for tipo in grupo_tipo.groups.keys():
    eh_tipo = dados['Tipo'] == tipo
    eh_dentro_limite = (dados['Valor'] >= limite_inferior[tipo]) & (dados['Valor'] <= limite_superior[tipo])
    selecao = eh_tipo & eh_dentro_limite
    dados_selecao = dados[selecao]
    dados_aluguel_v4 = pd.concat([dados_aluguel_v4, dados_selecao])
```

```python
dados_aluguel_v4.boxplot(['Valor'], by = ['Tipo'])
```
