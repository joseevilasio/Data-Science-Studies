```python
tipos_de_imoveis = dados["Tipo"]
tipos_de_imoveis.unique()
```

```python
tipos_de_imoveis.drop_duplicates(inplace=True)
tipos_de_imoveis
```

```python
tipos_de_imoveis = pd.DataFrame(tipos_de_imoveis)
tipos_de_imoveis.index = range(len(tipos_de_imoveis))
tipos_de_imoveis.columns.name = "index"
tipos_de_imoveis
```

```python
dados['Tipo'].value_counts()
tipos_de_imoveis_quantidade = pd.DataFrame(dados['Tipo'].value_counts()).reset_index()
tipos_de_imoveis_quantidade.columns=["Tipo Imovel", "Total"]
tipos_de_imoveis_quantidade
```

```python
residencial = ['Quitinete',
 'Casa',
 'Apartamento',
 'Casa de Condomínio',
 'Casa de Vila']
```

```python
dados.query(f"Tipo in {residencial}")
# # É possível fazer também utilizando: isin() passando como parametro a variavel 'residencial'
```

```python
dados_residencial = dados.query(f"Tipo in {residencial}")
dados_residencial = dados_residencial.reset_index().drop(columns="index")
dados_residencial
# É possível fazer um reset de index passando um range dentro de column.name
```
